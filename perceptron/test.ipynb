```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,10,0.1)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.arange(-5,5,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, linestyle='-', color='red', label='sin')
plt.plot(x,y2, linestyle='--', color='blue', label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin&Cos')
plt.legend()
plt.show()
```
